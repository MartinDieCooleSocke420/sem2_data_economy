```python
import requests


def print_response(response: requests.Response):
    print(f"Status code: {response.status_code}")
    print(f"Headers: {response.headers}")
    print(f"Body: {response.text}")
```

```python
response = requests.post(
    "http://localhost:1026/ngsi-ld/v1/entities",
    headers={"Content-Type": "application/ld+json"},
    json={
        "id": "urn:ngsi-ld:OffStreetParking:001",
        "type": "OffStreetParking",
        "name": {"type": "Property", "value": "Downtown Parking Lot"},
        "locationi": {
            "type": "GeoProperty",
            "value": {"type": "Point", "coordinates": [-3.7037902, 40.4167754]},
        },
        "category": {"type": "Property", "value": ["public"]},
        "totalCapacity": {"type": "Property", "value": 150},
        "availableCapacity": {"type": "Property", "value": 45},
        "@context": [
            "https://raw.githubusercontent.com/smart-data-models/dataModel.Parking/master/context.jsonld",
            "https://uri.etsi.org/ngsi-ld/v1/ngsi-ld-core-context-v1.6.jsonld",
        ],
    },
)
print_response(response)
```
